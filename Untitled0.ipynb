```python

crawl = app.crawl(
    "https://vocagame.com",
    limit=10
)

for page in crawl.data:
    print("TITLE:", page.metadata.title)
    print("CONTENT:", page.markdown[:300])
    print("--------------")
```

```python

crawl = app.crawl(
    "https://vocagame.com",
    limit=10
)

for page in crawl.data:
    print("TITLE:", page.metadata.title)

    text = page.markdown.split("\n")

    for line in text:
        if len(line) > 20:
            print(line)

    print("--------------")
```

```python

import json

data = []

for page in crawl.data:
    data.append({
        "title": page.metadata.title,
        "content": page.markdown
    })

with open("vocagame_data.json","w") as f:
    json.dump(data,f,indent=2)

print("Data berhasil disimpan")
```
